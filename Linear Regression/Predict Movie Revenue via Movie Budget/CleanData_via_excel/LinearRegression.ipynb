```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
data = pandas.read_csv('cost_revenue_clean.csv')
```

```python
data.describe()
```

```python
4.419e3
```

```python
X = DataFrame(data, columns=['production_budget_usd']) #Production budget as independent variable
y = DataFrame(data, columns=['worldwide_gross_usd']) #dependent variable
```

```python
plt.figure(figsize=(10,6)) #changing chart size
plt.scatter(X,y, alpha=0.3)  #create a scatter plot
plt.title("Film Cost vs Global Revenue")
plt.xlabel("Production Budget($)")
plt.ylabel("Worldwide Gross($)")
plt.xlim(0,450000000) #range size for x axis
plt.ylim(0,3000000000) #range size for y axis
plt.show()
```

```python
# h_theta(x)=theta_1(x) + theta_0 instead of y=mx+c
# where h stands for hypothesis
# residuals needs to be as less as possible using residual sum of squares

#linear regression
regression = LinearRegression()
regression.fit(X,y)
```

Our slope Coefficient is:

```python
regression.coef_ #theta_1
```

```python
#intercept
regression.intercept_  
```

```python
plt.figure(figsize=(10,6)) #changing chart size
plt.scatter(X,y, alpha=0.3)  #create a scatter plot
plt.plot(X,regression.predict(X), color='red', linewidth=4)
plt.title("Film Cost vs Global Revenue")
plt.xlabel("Production Budget($)")
plt.ylabel("Worldwide Gross($)")
plt.xlim(0,450000000) #range size for x axis
plt.ylim(0,3000000000) #range size for y axis
plt.show()
```

```python
#goodness of fit i.e R^2 or r^2
regression.score(X,y)  #amount of variation in movie revenue that is explained by the movie's budget
```

Here, a single feature i.e X=movie Budget can explain about 53%(goodness of fit) of variation in the movie's revenue.
But still theres a lot to learn as in does a movie's sequel affect its revenue or does it depend on its production time? So there are various other features we could work upon to analyze a movie's revenue.
